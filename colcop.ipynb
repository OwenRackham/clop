```python
from colorthief import ColorThief
color_thief = ColorThief('blondie.jpg')
# get the dominant color
dominant_color = color_thief.get_color(quality=1)
# build a color palette
palette = color_thief.get_palette(color_count=10)
palette
```

```python
from PIL import Image
import numpy

def closest_col( palette, col ):
    pal_length = len(palette)
    close_col = palette[0]
    close_dist = numpy.linalg.norm(numpy.array(col)-numpy.array(close_col))
    for x in range(1, pal_length):
        current_col = palette[x]
        current_dist = numpy.linalg.norm(numpy.array(col)-numpy.array(current_col))
        if current_dist < close_dist:
            close_col = current_col
    return close_col

img = Image.open('blondie.jpg')
pix = img.load()
dims = img.size

for x in range(1, dims[0]):
    for y in range(1, dims[1]):
        pix[x,y] = closest_col(palette,pix[x,y])
```
